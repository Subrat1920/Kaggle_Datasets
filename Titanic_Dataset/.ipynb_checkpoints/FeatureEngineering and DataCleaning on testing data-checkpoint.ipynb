# Feature Engineering and Data Cleaning on Testing Data

```python
import pandas as pd
import numpy as np
```

```python
test_data = pd.read_csv('test.csv')
test_data.head()
```

```python
test_data['cabin_multiple'] = test_data['Cabin'].apply(lambda x: 0 if pd.isna(x) else len(x.split(' ')))
test_data['cabin_multiple'].value_counts()
```

```python
## create categories based on the cabin letter (n stands for null)
## in this case we will treat null values like it's own category
test_data['cabin_adv'] = test_data['Cabin'].apply(lambda x : str(x)[0])
```

```python
## comparing survival rate by cabin
test_data.cabin_adv.value_counts()
```

```python
#understand ticket values better 
#numeric vs non numeric 
test_data['numeric_ticket'] = test_data.Ticket.apply(lambda x: 1 if x.isnumeric() else 0)
test_data['ticket_letters'] = test_data.Ticket.apply(lambda x: ''.join(x.split(' ')[:-1]).replace('.','').replace('/','').lower() if len(x.split(' ')[:-1]) >0 else 0)
```

```python
test_data['numeric_ticket'].value_counts()
```

```python
test_data['ticket_letters'].value_counts()
```

```python
test_data.Name.head(50)
```

```python
test_data['name_title'] = test_data.Name.apply(lambda x: x.split(',')[1].split('.')[0].strip())
```

```python
test_data.name_title.value_counts()
```

## Data Cleaning

```python
test_data.isnull().sum()
```

here we can drop 'Cabin', Fare has single null value so we can drop that row and we can convert the Age null values in mean values.

```python
mean_age = np.mean(test_data.Age)
test_data['Age'] = test_data['Age'].fillna(mean_age)
```

```python
test_data.dropna(subset=['Fare'], inplace=True)
```

Now, we can drop the un-necessary columns

```python
test_data.drop(columns=['Cabin', 'PassengerId', 'Name','Ticket'], inplace=True)
```

```python
test_data.isnull().sum()
```

So, we dont have any null value present in the dataset.

```python
test_data.head()
```

```python
## lets convert sibsp to log normal distribution
test_data['norm_sibsp'] = np.log(test_data.SibSp + 1)
test_data['norm_sibsp'].hist()
```

```python
## log norm for fare 
test_data['norm_fare'] = np.log(test_data.Fare + 1)
test_data.norm_fare.hist()
```

## Why do we used Log-Normalization Technique to the column
##### Applying a log-transformation to a column severs several purpose in data pre-processing and analysis:
    1. Reduce Skewness
    2. Handling Outliers
    3. Stabilizing Variance
    4. Improving Model Performance
    5. Interpretability
##### We add 1 to ensure that the log transformation works for row where column value is 0, as the logarithm of 0 in undefined.
